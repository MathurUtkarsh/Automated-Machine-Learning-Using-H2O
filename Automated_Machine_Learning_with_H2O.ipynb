```python
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

##**Data Preparation**

```python
df = h2o.import_file(path='https://www.kaggle.com/datasets/sahistapatel96/bankadditionalfullcsv?select=bank-additional-full.csv')
```

```python
df.head()
```

```python
df.describe(chunk_summary=True)
```

```python
train, test = df.split_frame(ratios=[0.8], seed = 1)
```

```python
train.columns
```

##**Train AutoML Models**

```python
aml = H2OAutoML(max_models =25,
                balance_classes=True,
		            seed =1
               )
```

```python
aml.train(training_frame = train, y = 'C1' )
```

##**Evaluation**

```python
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```
